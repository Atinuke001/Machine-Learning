Cleaning, formatting and fixing of archive data

```python
input_wiki = open('/Users/integrationalpha/Documents/tinu/EXT_TRANSACTION.csv', 'r',encoding = "ISO-8859-1")

cleaned_data = []

for i in input_wiki:
    x = i.split(";")
    if(len(x) == 38):
        new_x = x[0].split(',')
        first = True
        for ii in x:
            if(first):
                first = False
            else:
                new_x.append(ii)
        cleaned_data.append(new_x)
    elif(len(x) == 39):
        cleaned_data.append(x)
    else:
        print(len(x))
```

# Building the ML model

After much longer analysis of that type, you get an idea, how the data is and can then build an ML model to predict something with it, for example: Can I distinguish Cash and MT or can I predict typical sums out of the data. 

The algorithm is simple to implement. However, if the data is not processed like shown above, the results will be nonsense because for example the volumes will confuse the system, the exchanging rate would predict as datawhole always the Transaction country, etc....

From here on only examplifying code!!

To visualize still an example which predicts not much but shows the capabilities:

```python
import numpy as np
from sklearn.model_selection import train_test_split

transaction_CH_CASH = transaction_CH['Type']
transaction_CH_pred = transaction_CH.drop(['Type'],axis=1)
```

All the input has to be transformed like that: Following knowledge to decide if One-hot or labelencoding. Is 2,3,4,5 really better, if not, then use one-Hot encoding.

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
transaction_CH_pred['BENEFICIARY_COUNTRY'] = le.fit_transform(transaction_CH_pred['BENEFICIARY_COUNTRY'])
transaction_CH_pred['BENEFICIARY_BANK'] = le.fit_transform(transaction_CH_pred['BENEFICIARY_BANK'])
transaction_CH_pred['BENEFICIARY_COUNTRY'].head()
```

```python
transaction_CH_pred_used = transaction_CH_pred[['BENEFICIARY_COUNTRY','Volume','NUMBER_OF_SECURITIES','BENEFICIARY_BANK']]
transaction_CH_pred_used.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(transaction_CH_pred_used, transaction_CH_CASH, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(random_state=42)
```

```python
clf.fit(X_train, y_train)
```

And a simple prediction (accurancy)

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

The output indicates massive overfitting which is no wonder for such a simple demo. That means we will start to explore the model and its characteristica:
